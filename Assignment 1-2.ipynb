```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
```

```python
text = "NLTK is a versatile library for NLP tasks and includes functions fro word and sentence tokenization"
words = word_tokenize(text)
print(words)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(text)

tokens = [token.text for token in doc]
print(tokens)
```

```python
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
nltk_processed = [lemmatizer.lemmatize(token) for token in words if token.lower() not in stop_words]
```

```python
print(nltk_processed)
```

```python
from spacy import displacy
```

```python
text2 = "On February 17, 2025, Aidana Zhanabaeva and Arman Yermekov held a meeting at the Institute of Advanced Technologies and Data Analysis (IPTDA). The discussion focused on the development of new machine learning methods for predicting natural disasters."
doc = nlp(text2)
```

```python
for ent in doc.ents:
    print(ent.text, "->", ent.label_)

displacy.render(doc, style="ent", jupyter=True)
```

```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
inputs = tokenizer(text2, return_tensors="pt")
outputs = model(**inputs, output_hidden_states=True)

embeddings = outputs.last_hidden_state
print("Shape of the embeddings:", embeddings.shape)
```

```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
sentences = [
    "You did those products, and they work wonderful!",
    "This is the best event that I have experinced."
]
```

```python
print("Sentiment Analysis using Transformers:")
results = sentiment_pipeline(sentences)
for sentence, result in zip(sentences, results):
    print(f"Sentence: {sentence}\nPrediction: {result}\n")
```

```python
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

print("Sentiment Analysis using NLTK VADER:")
for sentence in sentences:
    scores = sia.polarity_scores(sentence)
    print(f"Sentence: {sentence}\nScores: {scores}\n")
```
